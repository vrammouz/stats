**Question 2.**  Explain whether each scenario is a classification or regression prob-
lem, and indicate whether we are most interested in inference or pre-
diction. Finally, provide n and p.

(a) We collect a set of data on the top 500 firms in the US. For each
firm we record profit, number of employees, industry and the
CEO salary. We are interested in understanding which factors
affect CEO salary.

(b) We are considering launching a new product and wish to know
whether it will be a success or a failure. We collect data on 20
similar products that were previously launched. For each prod-
uct we have recorded whether it was a success or failure, price
charged for the product, marketing budget, competition price,
and ten other variables.

(c) We are interested in predicting the % change in the USD/Euro
exchange rate in relation to the weekly changes in the world
stock markets. Hence we collect weekly data for all of 2012. For
each week we record the % change in the USD/Euro, the %
change in the US market, the % change in the British market,
and the % change in the German market.


---



**Answer** -> a) Since the outcome is numerical (salary), the problem is a regression problem. We would be interested in inference here because there is no sequence or pattern to learn but rather looking at the factors that affect the outcome variable.

*n* = 500 (firms), *p* = 3 (profit, employees, and industry)


b) The problem here requires a classification approach since there is a binary/categorical outcome variable (success or failure). We are interested in inference because there are several factors of interest at play here (price charged, marketing budget, competition price, and ten other variables).

*n* = 20 (products), *p* = 13


c) This is a regression problem and we would be particularly interested in prediction here since there are numerical patterns to be learned over time and the outcome is numerical.

*n* = 53 (weeks), *p* = 3 (US market, British market, and German market % changes)

**Question 5**. What are the advantages and disadvantages of a very flexible (versus
a less flexible) approach for regression or classification? Under what
circumstances might a more flexible approach be preferred to a less
flexible approach? When might a less flexible approach be preferred?


---



**Answer** -> A very flexible approach would lead to higher variance and less bias, yet more error. A less flexible approach would lead to more bias, less variance, but still high error rate (early training). A more flexible approach would be preferred when we need to accommodate for a large dataset with a lot of complexity whereas the opposite would be preferred when the data is simple and expected to have linear relationships.

**Question 6**. Describe the differences between a parametric and a non-parametric
statistical learning approach. What are the advantages of a para-
metric approach to regression or classification (as opposed to a non-
parametric approach)? What are its disadvantages?


---



**Answer** -> In parametric approaches, assumptions are made about how the data would fit the model, whereas non-parametric methods would estimate the predictions to be as close as possible to the data points without having a fixed form. The latter would require too many observations to obtain an accurate estimation of the predictions. The former would have assumptions about f that do not necessarily relfect the true f, whereas the latter avoids that. Parametric methods do allow for the capturing of significant relationships, however.
